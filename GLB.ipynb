```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/GLB.Ts+dSST_1.csv')
df['Decade']=df['Year']//10 *10
df1=df[::-1].reset_index()
```

```python
df=df[:-5]
df[df.columns[1:-1]]=df[df.columns[1:-1]].apply(pd.to_numeric)
df
```

```python
df_decades=[]
df_molten=[]
for i in df['Decade'].unique():
  df_i=df[df['Decade']==i].copy()
  df_decades.append(df_i)
  df2010s1=df_i.melt(id_vars=['Year','Decade'], value_vars=df_i.columns[1:-1])
  df_molten.append(df2010s1)
df['Decade'].unique()
```

```python
color_values=[]
for i in df_molten:
  color_values.append(i['value'].mean())

norm=mcolors.Normalize()
norm.autoscale(color_values)

color_rgba=sns.color_palette("coolwarm", as_cmap=True)(norm(color_values))
```

```python
fig, ax = plt.subplots(layout='constrained')

for i,j in enumerate(df_molten):
  sns.histplot(data=j, x="value", zorder=i, element='step', edgecolor='#000000', ax=ax, alpha=1, bins=11, stat='density', color=color_rgba[i])
ax.set(fc="#6ec5fa", title='The Average Global Temperature Keeps Climbing', ylabel='', xlabel='', ylim=ax.get_ylim(),xticks=[-1, -0.5, 0, 0.5, 1], xticklabels=['', '-0.5', '1951-1980 \n Average', '+0.5', '+1.0 deg C'])
df_1951_1980 = df[(df['Year']>=1951) & (df['Year']<=1980)].copy()
df_1951_1980=df_1951_1980.melt(id_vars=['Year','Decade'], value_vars=df_i.columns[1:-1])
ax.plot([df_1951_1980['value'].mean(), df_1951_1980['value'].mean()], ax.get_ylim(), color='#ffffff', zorder=len(df_molten), linestyle='--')
ax.tick_params(left = False, labelleft = False)

ax.text(-0.5, 4-0.17, 'pre-1950s', color='#ffffff')
ax.text(0.5, 4-0.5, '2000s', color='#ffffff')
ax.text(0.85, 4-1.5, '2010s', color='#ffffff')
```
